## Reading Data

> Reading Data

```python
fileName = 'data.json'
filePath = rootDir + '04. DataFrames/' + fileName

with open(filePath,'r') as f: 
    data = json.load(f)

    # reading grid
    grid = data['grid']

    # reading image path
    image_path  = data['image_path']
```

<br>

> Reading Sudoku Image

```python
# reading the Sudoku Puzzle
sudoku_img = cv2.imread(image_path)
```

<br>

## Reshaping the array into a 9 x 9 matrix

```python
# converting to numpy array
grid = np.array(grid)

# Reshaping the grid to a 9x9 matrix
puzzle = np.reshape(grid, (9,9))
```
